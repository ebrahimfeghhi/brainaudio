### B2T' 24: csv --> txt

```python
import pandas as pd

# --- Configuration ---
input_csv_path = '/home/lionelhu/brainaudio/results/test_files/transformer_24/neurips_b2t_24_bidir_transformer_seed_0__02_22_0512_test_3008380.csv' 
output_txt_path = '../results/b2t_24_test_results/neurips_b2t_24_bidir_transformer_seed_0.txt'
# ---------------------

# 1. Read the CSV
# We assume the file has a header. Based on your image, the text column is named 'text'.
df = pd.read_csv(input_csv_path)

# 2. Extract the 'text' column
# This automatically drops the 'id' column
sentences = df['text']

# 3. Save to .txt
# index=False removes the row numbers
# header=False removes the column name
sentences.to_csv(output_txt_path, index=False, header=False)

print(f"Successfully extracted {len(sentences)} sentences to {output_txt_path}")
```

### B2T' 25: txt --> csv

```python
import csv
# --- Configuration ---
input_txt_path = '/home/lionelhu/brainaudio/finetune_llm/generative_results/test_predictions_b2t_25_Meta-Llama-3.1-8B-Instruct-bnb-4bit_wfst_uni_seed_3416.txt' 
output_csv_path = '../finetune_llm/generative_results/test_predictions_b2t_25_Meta-Llama-3.1-8B-Instruct-bnb-4bit_wfst_uni_seed_3416.csv'
# ---------------------

with open(input_txt_path, 'r', encoding='utf-8') as infile, \
     open(output_csv_path, 'w', newline='', encoding='utf-8') as outfile:
    
    writer = csv.writer(outfile)
    writer.writerow(['id', 'text'])  # Write the header
    num_sentences = 0
    for idx, line in enumerate(infile):
        num_sentences += 1
        # Strip trailing newlines/whitespace and write the row
        clean_line = line.strip()
        if clean_line:  # Ignore empty lines just in case
            writer.writerow([idx, clean_line])


print(f"Successfully reformatted {num_sentences} sentences to {output_csv_path}")
```
