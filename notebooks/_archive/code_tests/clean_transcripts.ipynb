```python
import re
import pandas as pd
import pickle
```

```python
def clean_transcript(transcripts):
    cleaned_transcripts = []
    for transcript in transcripts:
        transcript = re.sub(r"[^a-zA-Z\- \']", "", transcript)
        transcript = transcript.replace("--", "").lower()
        cleaned_transcripts.append(transcript)
        
    return cleaned_transcripts
```

```python
def clean_transcript(transcripts):
    cleaned_transcripts = []
    for transcript in transcripts:
        # Handle case where transcript is a list (e.g., train data)
        if isinstance(transcript, list):
            # If it's a list, take the first element (assuming single-element lists)
            transcript = transcript[0] if len(transcript) > 0 else ""
        
        # Now clean the string
        transcript = re.sub(r"[^a-zA-Z\- \']", "", transcript)
        transcript = transcript.replace("--", "").lower()
        cleaned_transcripts.append(transcript)
        
    return cleaned_transcripts
```

```python
train_transcripts = pd.read_pickle("/home/ebrahim/data2/brain2text/b2t_24/transcripts_train.pkl")
```

```python
cleaned_transcript_24 = clean_transcript(train_transcripts)

save_file = "/home/ebrahim/data2/brain2text/b2t_24/transcripts_train_cleaned.pkl"
# 'wb' means "write bytes"
with open(save_file, mode="wb") as file:
    # This dumps the entire object into the file at once
    pickle.dump(cleaned_transcript_24, file)
```
