```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import pickle
```

```python
b2t24_dataset = pd.read_pickle("/data2/brain2text/b2t_24/brain2text24_with_fa")
b2t25_dataset = pd.read_pickle("/data2/brain2text/b2t_25/brain2text25_with_fa")
```

```python
# b2t_24 train trial stats
avg_trial_len_by_day_train = {}
total_len = 0
total_trial = 0

for idx in range(len(b2t24_dataset['train'])):
    day = b2t24_dataset['train'][idx]

    avg_trial_len_by_day_train[idx] = np.mean(day['timeSeriesLens'])
    
    for trial_len in day['timeSeriesLens']:
        total_len += trial_len 
        total_trial += 1

print(avg_trial_len_by_day_train)
print(f"Average length of all train trials of b2t_24 is {total_len / total_trial}")
```

```python
# b2t_24 val trial stats
avg_trial_len_by_day_val = {}
total_len = 0
total_trial = 0

for idx in range(len(b2t24_dataset['val'])):
    day = b2t24_dataset['val'][idx]

    avg_trial_len_by_day_val[idx] = np.mean(day['timeSeriesLens'])
    
    for trial_len in day['timeSeriesLens']:
        total_len += trial_len 
        total_trial += 1

print(avg_trial_len_by_day_val)
print(f"Average length of all val trials of b2t_24 is {total_len / total_trial}")
```

```python
# b2t_24 test trial stats
avg_trial_len_by_day_test = {}
total_len = 0
total_trial = 0

for idx in range(len(b2t24_dataset['test'])):
    day = b2t24_dataset['test'][idx]

    avg_trial_len_by_day_test[idx] = np.mean(day['timeSeriesLens'])
    
    for trial_len in day['timeSeriesLens']:
        total_len += trial_len 
        total_trial += 1

print(avg_trial_len_by_day_test)
print(f"Average length of all test trials of b2t_24 is {total_len / total_trial}")
```

```python
# b2t_25 train trial stats
avg_trial_len_by_day_train = {}
total_len = 0
total_trial = 0

for idx in range(len(b2t25_dataset['train'])):
    day = b2t25_dataset['train'][idx]

    avg_trial_len_by_day_train[idx] = np.mean(day['timeSeriesLen'])
    
    for trial_len in day['timeSeriesLen']:
        total_len += trial_len 
        total_trial += 1

print(avg_trial_len_by_day_train)
print(f"Average length of all train trials of b2t_25 is {total_len / total_trial}")
```

```python
# b2t_25 val trial stats
avg_trial_len_by_day_train = {}
total_len = 0
total_trial = 0

for idx in range(len(b2t25_dataset['val'])):
    day = b2t25_dataset['val'][idx]
    if day is None:
        avg_trial_len_by_day_train[idx] = None
    else:
        avg_trial_len_by_day_train[idx] = np.mean(day['timeSeriesLen'])
        for trial_len in day['timeSeriesLen']:
            total_len += trial_len  
            total_trial += 1

print(avg_trial_len_by_day_train)
print(f"Average length of all val trials of b2t_25 is {total_len / total_trial}")
```

```python
# b2t_25 test trial stats
avg_trial_len_by_day_test = {}
total_len = 0
total_trial = 0

for idx in range(len(b2t25_dataset['test'])):
    day = b2t25_dataset['test'][idx]
    if day is None:
        avg_trial_len_by_day_test[idx] = None
    else:
        avg_trial_len_by_day_test[idx] = np.mean(day['timeSeriesLen'])
        for trial_len in day['timeSeriesLen']:
            total_len += trial_len 
            total_trial += 1

print(avg_trial_len_by_day_test)
print(f"Average length of all test trials of b2t_25 is {total_len / total_trial}")
```
