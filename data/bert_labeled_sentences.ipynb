```python
threshold = 0.95                                                                                                                                                                         
input_file = '/data2/brain2text/c4_scored_sentences.tsv'                                                                                                                                     
output_file = f'/home/ebrahim/brainaudio/data/c4_filtered_sentences_{threshold}.txt'                                                                                                                      
                                                                                                                                                                                            
with open(input_file, 'r') as f_in, open(output_file, 'w') as f_out:                                                                                                                         
    next(f_in)  # skip header                                                                                                                                                                
    count = 0                                                                                                                                                                                
    for line in f_in:                                                                                                                                                                        
        parts = line.strip().split('\t', 1)                                                                                                                                                  
        if len(parts) == 2:                                                                                                                                                                  
            score = float(parts[0])                                                                                                                                                          
            if score > threshold:                                                                                                                                                            
                f_out.write(parts[1] + '\n')                                                                                                                                                 
                count += 1                                                                                                                                                                   
                                                                                                                                                                                            
print(f'Saved {count} sentences with score > {threshold} to {output_file}')     
```
