```python
  import json                                                                                                                                                                                          
                 
  with open("neurips_b2t_24_chunked_unidirectional_transformer_5to20_sec.json") as f:                                                                                                                  
      data = json.load(f)                                                                                                                                                                              

  print(f"{'Context Name':<25} | {'Mean Value':<12}")
  print("-" * 40)

  for context_name, values_dict in data.items():
      values = list(values_dict.values())
      if values:
          mean_val = sum(values) / len(values)
          print(f"{context_name:<25} | {mean_val:.6f}")
      else:
          print(f"{context_name:<25} | No data")
```

```python
# 1. Extract values for the two specific groups
from scipy import stats
group_7_5 = list(data["chunk_1_context_7.5"].values())
group_15 = list(data["chunk_1_context_15"].values())

# 2. Perform independent t-test
t_stat, p_value = stats.ttest_ind(group_7_5, group_15)

# 3. Print Results
print(f"Comparing Context 7.5 vs Context 15")
print(f"-----------------------------------")
print(f"T-statistic: {t_stat:.5f}")
print(f"P-value:     {p_value:.5f}")

if p_value < 0.05:
    print("Result: Statistically Significant (Reject Null Hypothesis)")
else:
    print("Result: Not Statistically Significant (Fail to Reject Null Hypothesis)")
```
