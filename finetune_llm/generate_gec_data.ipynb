```python
import json
from pathlib import Path

# ──────────────────────────────────────────────────────────────────
# 1) Constant Instruction
# ──────────────────────────────────────────────────────────────────
SYSTEM_PROMPT = (
    "Your task is to perform automatic speech recognition error correction. "
    "Below are multiple candidate transcriptions of the same utterance. "
    "These candidates were decoded from neural activity and may contain errors. "
    "Based on the candidates, produce the single most accurate, coherent, and grammatical "
    "transcription of the original utterance. Focus on key differences between candidates "
    "that change meaning or correctness, and avoid repetitive or nonsensical phrases. "
    "Respond with only the final corrected transcription—no explanations or extra text."
)

# ──────────────────────────────────────────────────────────────────
# 2) Conversion Function
# ──────────────────────────────────────────────────────────────────
def build_unsloth_jsonl(candidate_dict, labels_list, out_filepath, eval_mode=False):
    """
    Combines the candidates dictionary and ground-truth list into a JSONL file.
    
    Args:
        candidate_dict (dict): Dictionary mapping string indices (e.g., '0') to lists of candidates.
        labels_list (list): List of ground truth strings.
        out_filepath (str): Destination path for the .jsonl file.
        eval_mode (bool): If True, leaves the assistant completion empty.
    """
    out_path = Path(out_filepath)
    out_path.parent.mkdir(parents=True, exist_ok=True)
    
    # Sort keys as integers to ensure we iterate in the correct order 0, 1, 2...
    sorted_keys = sorted(candidate_dict.keys(), key=lambda x: int(x))
    
    with out_path.open("w", encoding="utf-8") as fout:
        for key in sorted_keys:
            idx = int(key)
            candidates = candidate_dict[key]
            
            # 1. Build candidates text block (10 lines usually)
            cand_trans = "\n".join(
                f"{i+1}) {cand}" for i, cand in enumerate(candidates)
            )
            
            # 2. Construct the user prompt
            prompt_content = f"{SYSTEM_PROMPT}\n\nCandidates:\n{cand_trans}"
            
            # 3. Construct the completion 
            # (Check bounds just in case there's a mismatch between dict and list lengths)
            if idx < len(labels_list):
                gt_sentence = labels_list[idx].strip()
            else:
                gt_sentence = ""
                
            completion_content = "" if eval_mode else gt_sentence

            # 4. Build Unsloth/LLaMA-3 compatible JSON object
            obj = {
                "prompt": [
                    {
                        "content": prompt_content,
                        "role": "user"
                    }
                ],
                "completion": [
                    {
                        "content": completion_content,
                        "role": "assistant"
                    }
                ]
            }
            
            fout.write(json.dumps(obj, ensure_ascii=False) + "\n")
            
    print(f"✅ Wrote {len(sorted_keys)} examples to {out_path.resolve()}")
```

```python
# Assuming test_data_24 and test_labels are loaded
import json
from pathlib import Path

SYSTEM_PROMPT = (
    "Your task is to perform automatic speech recognition error correction. "
    "Below are multiple candidate transcriptions of the same utterance. "
    "These candidates were decoded from neural activity and may contain errors. "
    "Based on the candidates, produce the single most accurate, coherent, and grammatical "
    "transcription of the original utterance. Focus on key differences between candidates "
    "that change meaning or correctness, and avoid repetitive or nonsensical phrases. "
    "Respond with only the final corrected transcription—no explanations or extra text."
)

def build_test_jsonl(candidate_dict, out_filepath):
    """
    Formats candidate beams into a test JSONL without requiring ground truth labels.
    """
    out_path = Path(out_filepath)
    out_path.parent.mkdir(parents=True, exist_ok=True)
    
    # Sort keys as integers to maintain order (0, 1, 2...)
    sorted_keys = sorted(candidate_dict.keys(), key=lambda x: int(x))
    
    with out_path.open("w", encoding="utf-8") as fout:
        for key in sorted_keys:
            candidates = candidate_dict[key]
            
            # Format the 10 candidate beams
            cand_trans = "\n".join(f"{i+1}) {cand}" for i, cand in enumerate(candidates))
            prompt_content = f"{SYSTEM_PROMPT}\n\nCandidates:\n{cand_trans}"
            
            # For inference, the assistant completion is empty
            obj = {
                "prompt": [{"content": prompt_content, "role": "user"}],
                "completion": [{"content": "", "role": "assistant"}]
            }
            
            fout.write(json.dumps(obj, ensure_ascii=False) + "\n")
            
    print(f"✅ Wrote {len(sorted_keys)} test examples to {out_path.resolve()}")
```
