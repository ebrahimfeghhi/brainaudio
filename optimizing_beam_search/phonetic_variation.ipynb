```python
import nltk
nltk.download('cmudict')
from nltk.corpus import cmudict

# Load the dictionary once (it's a hash map)
d = cmudict.dict()

def get_variants(word):
    # Returns a list of lists, e.g., [['F', 'AY1', 'R'], ['F', 'AY1', 'ER0']]
    return d.get(word.lower(), []) 

print(get_variants("america"))
```
